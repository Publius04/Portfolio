```python
import pandas as pd
import plotly.express as px
import os
import json
```

```python
pwd = os.getcwd()
pwd
```

### Import Variants Data

```python
with open("variant.json", "r") as f:
    variants = pd.DataFrame(json.load(f)["USA"])
variants["week"] = pd.to_datetime(variants["week"], format=r"%Y-%m-%d")
variants
```

### CLEANING

```python
weeks = variants["week"]
variant_names = variants.columns[1:]
```

```python
week = []
variant = []
value = []

for name in variant_names:
    week += list(weeks)
    variant += [name] * len(weeks)
    value += list(variants[name])

vdf = pd.DataFrame({"week": week, "variant": variant, "value": value})
vdf
```

### Graph of Alpha over time

```python
fig = px.line(data_frame=variants, x="week", y="20I (Alpha, V1)", line_shape="spline")
fig.show()
```
