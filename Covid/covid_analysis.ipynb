```python
import pandas as pd
import plotly.express as px
import os
```

```python
pwd = os.getcwd()
```

```python
df = pd.read_csv(pwd + "\\covidstate.csv")
df["submission_date"] = pd.to_datetime(df["submission_date"])
df.sort_values(by=["state", "submission_date"], inplace=True)
df.reset_index(inplace=True)
df.drop(columns="index", inplace=True)
```

### Cleaning for Graph of AZ

```python
df.columns
```

```python
az_df = df[df["state"] == "AZ"][["submission_date", "new_case"]]
```

```python
fig = px.scatter(az_df, x="submission_date", y="new_case")
fig.show()
```
